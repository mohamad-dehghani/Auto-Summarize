```python
def getTextWaPo(url):
    page = urllib.request.urlopen(articleURL).read().decode('utf8','ignore')
    soup = BeautifulSoup(page,"html.parser")
    for script in soup(["script", "style"]):                   
        script.decompose()
    text = ' '.join(map(lambda p: p.text.strip(), soup.find_all("div", {"class": "entry-content clearfix single-post-content"})))
    return text
```

```python
text = getTextWaPo(articleURL)
```

```python
sents = sent_tokenize(text)
sents[:5]
```

```python
word_sent = word_tokenize(text)
word_sent[:5]
```

```python
stopwords = []
file = open('stopwords.txt', encoding = 'utf-8').read()
[stopwords.append(x) for x in file.split()]
_stopwords = set(stopwords)
```

```python
word_sent = [word for word in word_sent if word not in _stopwords]
```

```python
word_sent[:5]
```

```python
freq = FreqDist(word_sent)
freq
```

```python
ranking = defaultdict(int)
for i,sent in enumerate(sents):
    for w in word_tokenize(sent):
        if w in freq:
            ranking[i] += freq[w]
```

```python
sents_idx = nlargest(3, ranking, key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```

```python
def summarize(text, n):
    sents = sent_tokenize(text)
    
    assert n <= len(sents)
    word_sent = word_tokenize(text)    
    word_sent=[word for word in word_sent if word not in _stopwords]
    freq = FreqDist(word_sent)
    
    ranking = defaultdict(int)
    
    for i,sent in enumerate(sents):
        for w in word_tokenize(sent):
            if w in freq:
                ranking[i] += freq[w]
        
    sents_idx = nlargest(n, ranking, key=ranking.get)
    return [sents[j] for j in sorted(sents_idx)]
```

```python
res = summarize(text,3)
for sent in res:
    print(sent)
```
